```python

N=20
# x = np.linspace(-1, 1, N)
# y = np.linspace(-1, 1, N)
dx=1/N
i = np.arange(N)
j = np.arange(N)

x = (i - N/2) * dx
y = (j - N/2) * dx

X, Y = np.meshgrid(x, y)

V = 4. * (X**2 + Y**2)

plt.figure()
plt.imshow(V.T, origin='lower', extent=(-1, 1, -1, 1))
plt.colorbar()
plt.title('Harmonic Potential V(x,y)')
plt.savefig("visualize_harmonic_potential")
plt.show()
```
