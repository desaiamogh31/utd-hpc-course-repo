```python
import numpy as np
import matplotlib.pyplot as plt
#Define the directory where your data files are located
data_dir = "/Users/a.n.d./Documents/UTD/Coursework/HPC/Assignments/HW2/results_local"

N = [10, 20, 40, 80, 120, 160, 200, 240, 320] # Adjust as needed (use the largest N you ran on HPC)

#To visualize the eigenvalues for different N values, we can plot them on the same graph for comparison.
for n in N:
    data = np.loadtxt(f'{data_dir}/eigs_N{n}_well.txt')
    plt.plot(data, 'o-', label=f'N={n}')
    
plt.xlabel('Eigenvalue Index')
plt.ylabel('Energy (arbitrary units)')
plt.title(f'Eigenvalues(Well) from HPC, varying N')
plt.legend()
plt.savefig(f"{data_dir}/eigenvalues_N_well.png")
plt.show()
plt.close()

```

```python
# To visualize the eigenvalues for the harmonic potential case, we can do the same as above but with the files corresponding to the harmonic potential.
for n in N:
    data = np.loadtxt(f'{data_dir}/eigs_N{n}_harmonic.txt')
    plt.plot(data, 'o-', label=f'N={n}')

plt.xlabel('Eigenvalue Index')
plt.ylabel('Energy (arbitrary units)')
plt.title(f'Eigenvalues(Harmonic) from HPC, varying N')
plt.legend()
plt.savefig(f"{data_dir}/eigenvalues_N_harmonic.png")
plt.show()
```

```python
# Now, let's analyze the convergence of the ground state energy as we increase N. 
E0 = []

for n in N:
    vals = np.loadtxt(f"{data_dir}/eigs_N{n}_well.txt")
    E0.append(vals[0])   # ground state energy

plt.plot(N, E0, 'o-')
plt.xlabel("N")
plt.ylabel("Ground state energy")
plt.title("Grid convergence (well)")
plt.savefig(f"{data_dir}/grid_convergence_well.png")
plt.show()
```

```python
# Now, let's analyze the convergence of the ground state energy as we increase N. 
E0 = []
E1 = []
E2 = []
E3 = []
E4 = []
E5 = []

for n in N:
    vals = np.loadtxt(f"{data_dir}/eigs_N{n}_well.txt")
    E0.append(vals[0])   # ground state energy
    E1.append(vals[1])   # first excited state energy
    E2.append(vals[2])   # second excited state energy
    E3.append(vals[3])   # third excited state energy
    E4.append(vals[4])   # fourth excited state energy
    E5.append(vals[5])   # fifth excited state energy

plt.plot(N, E0, 'o-', label="Ground state")
plt.plot(N, E1, 'o-', label="First excited state")
plt.plot(N, E2, 'o-', label="Second excited state")
plt.plot(N, E3, 'o-', label="Third excited state")
plt.plot(N, E4, 'o-', label="Fourth excited state")
plt.plot(N, E5, 'o-', label="Fifth excited state")

plt.xlabel("N")
plt.ylabel("Energy (arbitrary units)")
plt.title("Grid convergence (well)")
plt.legend()
plt.savefig(f"{data_dir}/grid_convergence_well.png")
plt.show()
```

```python
# Now, let's analyze the convergence of the ground state energy as we increase N. 
E0 = []
E1 = []
E2 = []
E3 = []
E4 = []
E5 = []

for n in N:
    vals = np.loadtxt(f"{data_dir}/eigs_N{n}_harmonic.txt")
    E0.append(vals[0])   # ground state energy
    E1.append(vals[1])   # first excited state energy
    E2.append(vals[2])   # second excited state energy
    E3.append(vals[3])   # third excited state energy
    E4.append(vals[4])   # fourth excited state energy
    E5.append(vals[5])   # fifth excited state energy

plt.plot(N, E0, 'o-', label="Ground state")
plt.plot(N, E1, 'o-', label="First excited state")
plt.plot(N, E2, 'o-', label="Second excited state")
plt.plot(N, E3, 'o-', label="Third excited state")
plt.plot(N, E4, 'o-', label="Fourth excited state")
plt.plot(N, E5, 'o-', label="Fifth excited state")

plt.xlabel("N")
plt.ylabel("Energy (arbitrary units)")
plt.title("Grid convergence (harmonic)")
plt.legend()
plt.savefig(f"{data_dir}/grid_convergence_harmonic.png")
plt.show()
```
