```python
import numpy as np
import matplotlib.pyplot as plt
# Load HPC results
N = [10, 20, 40, 80] # Adjust as needed (use the largest N you ran on HPC)

#To visualize the eigenvalues for different N values, we can plot them on the same graph for comparison.

for n in N:
    data = np.loadtxt(f'eigs_N{n}_well.txt')
    plt.plot(data, 'o-', label=f'N={n}')
    
plt.xlabel('Eigenvalue Index')
plt.ylabel('Energy (arbitrary units)')
plt.title(f'Eigenvalues from HPC, varying N')
plt.legend()
plt.savefig(f"eigenvalues_N{N}_well")
plt.show()
plt.close()

```

```python
# To visualize the eigenvalues for the harmonic potential case, we can do the same as above but with the files corresponding to the harmonic potential.
for n in N:
    data = np.loadtxt(f'eigs_N{n}_harmonic.txt')
    plt.plot(data, 'o-', label=f'N={n}')

plt.xlabel('Eigenvalue Index')
plt.ylabel('Energy (arbitrary units)')
plt.title(f'Eigenvalues from HPC, varying N')
plt.legend()
plt.savefig(f"eigenvalues_N{N}_harmonic")
plt.show()
```

```python
# Now, let's analyze the convergence of the ground state energy as we increase N. 
E0 = []

for n in N:
    vals = np.loadtxt(f"eigs_N{n}_well.txt")
    E0.append(vals[0])   # ground state energy

plt.plot(N, E0, 'o-')
plt.xlabel("N")
plt.ylabel("Ground state energy")
plt.title("Grid convergence (well)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

E0 = []
E1= []
E2 =[]
E3 = []
E4 = []
for n in N:
    vals = np.loadtxt(f"eigs_N{n}_harmonic.txt")
    E0.append(vals[0])   # ground state energy
    E1.append(vals[1])   # first excited state energy
    E2.append(vals[2])   # second excited state energy
    E3.append(vals[3])   # third excited state energy
    E4.append(vals[4])   # fourth excited state energy
for n in N:
    vals = np.loadtxt(f"eigs_N{n}_harmonic.txt")
    



plt.plot(N, E0, 'o-', label='Ground state')
plt.plot(N, E1, 'o-', markersize=8,linewidth =4, label='First excited state')
plt.plot(N, E2, '*-', label='Second excited state')
plt.plot(N, E3, 'o-', label='Third excited state')
plt.plot(N, E4, 'o-', label='Fourth excited state')
plt.xlabel("N")
plt.ylabel("Energy")
plt.title("Grid convergence (harmonic)")
plt.legend()
plt.show()
```

```python

N=20
# x = np.linspace(-1, 1, N)
# y = np.linspace(-1, 1, N)
dx=1/N
i = np.arange(N)
j = np.arange(N)

x = (i - N/2) * dx
y = (j - N/2) * dx

X, Y = np.meshgrid(x, y)

V = 4. * (X**2 + Y**2)

plt.figure()
plt.imshow(V.T, origin='lower', extent=(-1, 1, -1, 1))
plt.colorbar()
plt.title('Harmonic Potential V(x,y)')
plt.savefig("visualize_harmonic_potential")
plt.show()
```
