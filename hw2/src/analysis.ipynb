```python
# Ground/Excited-state convergence from TXT files (indices 0..6)
import numpy as np
import matplotlib.pyplot as plt

data_dir = "/Users/a.n.d./Documents/UTD/Coursework/HPC/Assignments/HW2/results_local"
potential_tag = "anisotropic_harmonic"
N_values = np.array([10, 20, 40, 80, 160, 320], dtype=int)

state_indices = range(0, 7)  # 0,1,2,3,4,5,6
energies = {k: [] for k in state_indices}

# Read eigenvalues for each N
for n in N_values:
    file_path = f"{data_dir}/eigs_N{n}_{potential_tag}.txt"
    vals = np.atleast_1d(np.loadtxt(file_path))
    for k in state_indices:
        energies[k].append(float(vals[k]))

# Convert to arrays and compute consecutive differences
for k in state_indices:
    energies[k] = np.array(energies[k], dtype=float)

N_diff = 2 * np.diff(N_values)  # keep your original definition

# Fit + plot for each state index
for k in state_indices:
    E_diff = np.diff(energies[k])

    # log-fit only where positive
    mask = E_diff > 0
    x = N_diff[mask]
    y = E_diff[mask]

    if len(x) < 2:
        print(f"Index {k}: not enough positive points for log-log fit.")
        continue

    m, b = np.polyfit(np.log(x), np.log(y), 1)
    y_fit = np.exp(b) * (x ** m)

    plt.figure(figsize=(6, 5))
    plt.loglog(x, y, "o", label=f"State index {k} diff")
    plt.loglog(x, y_fit, "-", label=f"Fit: log(y)={m:.4f} log(N)+{b:.4f}")
    plt.xlabel("N")
    plt.ylabel(rf"Energy difference $\Delta E_{{{k}}}$")
    plt.title(f"Convergence for eigenvalue index {k} ({potential_tag})")
    plt.legend()
    plt.tight_layout()
    plt.show()

    print(f"Index {k}: log(ΔE) = {m:.6f} * log(N) + {b:.6f}")
    print(f"Index {k}: ΔE = exp({b:.6f}) * N^({m:.6f})")
```
